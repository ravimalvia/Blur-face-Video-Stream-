## Using HaarCascade Classifier

```python
#using Haarcascade Classifier

#import computer vision module
import cv2

#Capture Video using Default Camera
Cap=cv2.VideoCapture(0)

#for TEXT Properties
font=cv2.FONT_HERSHEY_SIMPLEX
font_scale=1
thinkness=2
color1= ( 255, 255, 255) # White Color
color2= ( 0, 0, 255) #Red color

 
#Classifier
classifier= cv2.CascadeClassifier("C:/Users/MALVI/OneDrive/Desktop/Arth/DataScience/haarcascade_frontalface_alt.xml")


while True:
    ret, Image =Cap.read()
    face=classifier.detectMultiScale(Image)
    if face == ():
            text="No Face is detected"
            cv2.putText(Image, text, (50, 50) ,font, font_scale, color1, thinkness )
            cv2.imshow("live Capture", Image)
            if cv2.waitKey(1) == 27:  #ESC key
                break
    else: 
            text="Face detected"
            print(text)
            for (x,y,w,h) in face:
                start_point=(x,y)
                end_point=(x+w,y+h)
                cv2.rectangle(Image, start_point, end_point ,(255,0, 0))
                cv2.putText(Image, text, (50, 50) ,font, font_scale,color2 , thinkness )
                face_cut=Image[y:y+h, x:x+w]
                blur_face=cv2.blur(face_cut, ksize=(50,50)) # ksize = kernel Size
                Image[y:y+h, x:x+w]=blur_face
            cv2.imshow("live Capture", Image)
            if cv2.waitKey(1) == 27: #Esc Key
                    break

cv2.destroyAllWindows()
Cap.release()
```
